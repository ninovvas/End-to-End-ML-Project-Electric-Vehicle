If we can see there are 170 countries. Now is the question, how we can transform the countries to a number as categories? If we use the __LabelEncoder__ provided by the sklearn library, so
all countries will be assign to unique numerical label in the feature. The problem is that __LabelEncoder__ is appropriate when there is an ordinal relationship between the categories (i.e., one category is greater than or less than another). In our case this can be a problem for some of the machine algorithms, because using __LabelEncoder__ might introduce unintended relationships in the model. 
Another option is to use __OneHotEncoder__. OneHotEncoder is used to convert categorical integer features into one-hot encoded vectors. Each integer value is represented as a binary vector where all values are zero except the index that is marked with a 1. In our case we will inscrease the features with additiolan 170 features. This can be a problem, because of the heigh varians by the learning the model.
So this two solutions maybe are not the correct solution for the column __Country__. So what we want to acheaved? Mainly we want to present the rank of each country base on the saled vehicles in this country. This will be provide us the posibility to clarify how importent is each country and to set the priority of each country.

```python
scaler = MinMaxScaler()
data_rescaled = scaler.fit_transform(electric_vehicles_attributes_train)
number_of_train_features = electric_vehicles_attributes_train.shape[1]

pca = PCA().fit(data_rescaled)

plt.rcParams["figure.figsize"] = (30,8)

fig, ax = plt.subplots()
xi = np.arange(1, number_of_train_features  + 1, step=1)
y = np.cumsum(pca.explained_variance_ratio_)

plt.ylim(0.0,1.1)
plt.plot(xi, y, marker='o', linestyle='--', color='b')

plt.xlabel('Number of Components')
plt.xticks(np.arange(0, number_of_train_features , step=15))
plt.ylabel('Cumulative variance (%)')
plt.title('The number of components needed to explain variance')

plt.axhline(y=0.98, color='r', linestyle='-')
plt.text(0.5, 0.85, '98% cut-off threshold', color = 'red', fontsize=16)

ax.grid(axis='x')
plt.show()
```

Each row of the matrix represents the instances in a predicted class, while each column represents the instances in an actual class (or vice versa). The numbers in the matrix represent the counts of instances. Let's break down the interpretation:

Row 1 (Class 0): All instances in this class are predicted as Class 3 (the fourth column), as the counts are in the last column (596 instances).

Row 2 (Class 1): Similarly, all instances in this class are predicted as Class 3 (the fourth column), as the counts are in the last column (286 + 399 instances).

Row 3 (Class 2): The majority of instances in this class are correctly predicted as Class 3 (the fourth column), but there are some instances predicted as Class 4 (the fifth column).

Row 4 (Class 3): The majority of instances in this class are correctly predicted as Class 3 (the fourth column), but there are some instances predicted as Class 2 (the third column).

Row 5 (Class 4): All instances in this class are predicted as Class 3 (the fourth column), as the counts are in the last column (18 + 921 instances).

Here's how to interpret it in terms of standard confusion matrix notation:

True Positives (TP): Instances correctly predicted as the respective class.

True Negatives (TN): Instances correctly predicted as not being in the respective class (there are none in this case).

False Positives (FP): Instances incorrectly predicted as being in the respective class.

False Negatives (FN): Instances incorrectly predicted as not being in the respective class.

In this specific confusion matrix:

Class 0 (Row 1): All instances are incorrectly predicted as Class 3, so TP = 0, FP = 596, TN = 0, FN = 0.
Class 1 (Row 2): All instances are incorrectly predicted as Class 3, so TP = 0, FP = 286 + 399, TN = 0, FN = 0.
Class 2 (Row 3): Mostly correctly predicted as Class 3, with some instances incorrectly predicted as Class 4.
Class 3 (Row 4): Mostly correctly predicted as Class 3, with some instances incorrectly predicted as Class 2.
Class 4 (Row 5): All instances are incorrectly predicted as Class 3, so TP = 0, FP = 18 + 921, TN = 0, FN = 0.
